```python
import pandas as pd
area='Shikoku'
df = pd.read_csv(f'tso_{area}.csv')
df
```

```python
# area, datetime 以外は全て数値にする
for col in df.columns:
    if col not in ['area', 'datetime']:
        df[col] = pd.to_numeric(df[col], errors='coerce')
df
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df
```

```python
df.to_csv(f'tso_{area}.csv', index=False)
```
